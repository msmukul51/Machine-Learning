```python
def convert2(obj):#this function is extracting first 3 name of the actors
  l=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      l.append(i['name'])
      counter+=1
    else:
      break
  return l

```

else:

return l

```python
def extract_director_name(obj):#this function extract the name of the director out of the unfiltered data
  l=[]
  for i in ast.literal_eval (obj):
    if i['job']=='Director':
      l.append(i['name'])
  return l
```

```python
movies_df['crew']=movies_df['crew'].apply(extract_director_name)
```

```python
#[Action, Adventure, Fantasy, Science Fiction] CODE FOR REMOVING THE SPACE BETWEEN SCIENCE AND FICTION
movies_df['genres']=movies_df['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies_df['keywords']=movies_df['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies_df['cast']=movies_df['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies_df['crew']=movies_df['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
#here in columns there are some double commas code for removing them
# Assuming you have already loaded your DataFrame named 'movies_df'

def remove_double_commas(lst):
    return [item.replace(',,', ',') for item in lst]

# Apply the function to the specified columns
columns_to_clean = ['genres', 'keywords', 'cast', 'crew','overview']

for column in columns_to_clean:
    movies_df[column] = movies_df[column].apply(remove_double_commas)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```
